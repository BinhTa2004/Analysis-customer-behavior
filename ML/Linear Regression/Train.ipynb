```python
import numpy as np
from Model import LinearRegression
from matplotlib import pyplot as plt
model = LinearRegression()
```

```python
new_heights = np.array([[163], [170], [183]])
predicted_weights = model.predict(new_heights)

# In kết quả dự đoán
for h, w in zip(new_heights.flatten(), predicted_weights.flatten()):
    print(f"Heigh: {h} cm -> Predict Weight: {w:.2f} kg")
```

```python
X = np.array([[147, 150, 153, 158, 163, 165, 168, 170, 173, 175, 178, 180, 183]]).T
y = np.array([[49,  50,   51,  54,  58,  59,  60,  62,  63,  64,  66,  67, 68]]).T

model.fit(X, y)

y_pred = model.predict(X)

print("Weight w (regression coefficient):")
print(model.W)
```

```python
plt.plot(X, y, 'ro', label='Data original')
plt.plot(X, y_pred, label='Line regression')
plt.axis([140, 190, 45, 75])
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.legend()
plt.show()
```
