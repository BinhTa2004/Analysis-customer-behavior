```python
from sklearn.datasets import load_sample_image
import numpy as np
import matplotlib.pyplot as plt

# Load sample image and convert to grayscale
original_image = load_sample_image("china.jpg")
gray_image = np.mean(original_image, axis=2)  # Convert to grayscale

# Display the original grayscale image
plt.imshow(gray_image, cmap='gray')
plt.title("Original Grayscale Image")
plt.show()

# Show the shape of the grayscale image
print("Grayscale Image Shape:", gray_image.shape)
```

```python
# Define the block size (28x28 pixels)
block_size = 28
samples = []
labels = []

# Divide the image into 28x28 blocks
for i in range(0, gray_image.shape[0] - block_size + 1, block_size):
    for j in range(0, gray_image.shape[1] - block_size + 1, block_size):
        block = gray_image[i:i + block_size, j:j + block_size].flatten()
        samples.append(block)
        labels.append(0)  # Placeholder label; you can replace it if needed

# Convert to numpy arrays
X_train = np.array(samples)
Y_train = np.array(labels)

# Show the shape of X_train and Y_train
print("X_train Shape:", X_train.shape)
print("Y_train Shape:", Y_train.shape)
print("Example Flattened Block (First Sample):\n", X_train[0])
```

```python
import pandas as pd

# Create a dictionary to hold label and pixel data
pixel_data = {"label": Y_train}
for i in range(X_train.shape[1]):
    pixel_data[f"pixel {i+1}"] = X_train[:, i]

# Convert to DataFrame
df_train = pd.DataFrame(pixel_data)

# Display the first few rows of the DataFrame
df_train.head()
```

```python
# Display a sample 28x28 block as an image to verify reshaping
sample_block = X_train[0].reshape((block_size, block_size))
plt.imshow(sample_block, cmap='gray')
plt.title("Sample 28x28 Block (Flattened)")
plt.show()
```
